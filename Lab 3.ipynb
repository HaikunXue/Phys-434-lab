#### A)   
- Statistical Question:  
If an observation is made on a stationary star,   
we expect the observation distribution to be a Gaussian Distribution  
with a RMS of 1 arcsec. How far does the signal need to move to produce  
a 5 $\sigma$ detection?

- Math Equation:  
The movement would be a Rayleigh Distribution:  
$$
 f(x, \sigma )={\frac {x}{\sigma ^{2}}}e^{-x^{2}/(2\sigma ^{2})}
$$  
Then the distance n would satisfy:  
$$
P(5\sigma_g) = \int^{+\infty}_{n} pdf_{Ray}(x, \sigma_r) dx
$$  
Where $\sigma_r = 1$ arcsec, $P (5 \sigma_g) \approx 1$ in 3.5 million

- Then: 

```python
prob = stats.norm.cdf(5)
stats.rayleigh.ppf(prob)
```

The distance would be 5.489 arcsecs

- Statistical Question:  
If the area on the sky is not being influenced by the Moon,  
we expect the distribution to be Gaussian with a mean of 1 cosmic ray  
per minute. What is the probability of observing a sum   
of 6800 cosmic rays over 7200 minutes?

- Math Equation:  
It would be a Poisson Distribution with mean $\mu =  7200$
$$
P = \int^{+\infty}_{6800} pdf_{Poi}(x, \mu) dx
$$

- Calculation:

```python
prob = stats.poisson.cdf(6800,7200)
prob
```

```python
sigma = stats.norm.ppf(prob)
sigma
```

It would be 4.75 $\sigma$
