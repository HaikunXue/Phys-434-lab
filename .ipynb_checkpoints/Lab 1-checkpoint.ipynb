### A Little Statistic

```python
a = 2.12
signal_Value = 5.342
d = stats.gamma.rvs(a, size = 100000)
```

```python
fig, ax = plt.subplots(1, 1)

ax.hist(d,50, density=True)
plt.tick_params(labelsize = 24)
plt.xlim([0,7])
plt.ylim([0, .38])
plt.title("Gamma Distribution")

x = np.linspace(0,10,100000)
y = np.linspace(0,0.5,1000)
ax.plot(np.ones_like(y)*signal_Value,y,linewidth = 5, linestyle = ':', color = 'red')
ax.plot(x,stats.gamma.pdf(x,a),linewidth = 8,alpha = 0.7)

plt.show()
```

### Non-continuous distributions

```python
fig, ax = plt.subplots(2, 2)
x = np.linspace(-5,5,100000)
mu = [3,9,27,81]

x = np.arange(stats.poisson.ppf(0.01, mu[0]),stats.poisson.ppf(0.99, mu[0]))
ax[0,0].step(x, stats.poisson.pmf(x, mu[0]), where='mid',linewidth = 4,alpha = 0.7,color = "orange")
ax[0,0].set_title("Poisson Distribution, mu = 3")

x = np.arange(stats.poisson.ppf(0.01, mu[1]),stats.poisson.ppf(0.99, mu[1]))
ax[0,1].step(x, stats.poisson.pmf(x, mu[1]), where='mid',linewidth = 4,alpha = 0.7,color = "orange")
ax[0,1].set_title("Poisson Distribution, mu = 9")

x = np.arange(stats.poisson.ppf(0.01, mu[2]),stats.poisson.ppf(0.99, mu[2]))
ax[1,0].step(x, stats.poisson.pmf(x, mu[2]), where='mid',linewidth = 4,alpha = 0.7,color = "orange")
ax[1,0].set_title("Poisson Distribution, mu = 27")

x = np.arange(stats.poisson.ppf(0.01, mu[3]),stats.poisson.ppf(0.99, mu[3]))
ax[1,1].step(x, stats.poisson.pmf(x, mu[3]), where='mid',linewidth = 4,alpha = 0.7,color = "orange")
ax[1,1].set_title("Poisson Distribution, mu = 81")

plt.show()
```

As the number of the value increases, the Poisson distribution approaches a normal distribution

If we assume the background noise is a Poisson distribution, with $\mu = 3$, what would be probability be to observe a value greater than  5.342?

```python
p_Poi = stats.poisson.cdf(signal_Value, mu[0])
```

```python
stats.norm.ppf(p_Poi)
```

It will be 1.38 $\sigma$
