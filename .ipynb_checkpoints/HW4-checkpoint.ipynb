```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

### Problem 1.

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size= int(1e5))*np.sqrt(N)+N
```

```python
t_1k = t[:1000 + 1]
bkgd_1k = bkgd[:1000 + 1]

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

ax.scatter(t_1k, bkgd_1k, color = "orange")
ax.set_title('1a) First {} points of random data'.format(1000), fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background Distribution Value", fontsize = 20);
```

```python
#initial and final value for contiguous points
t_Ini = 849
t_Fin = 2349

t_Par = t[t_Ini:t_Fin + 1]
bkgd_Par = bkgd[t_Ini:t_Fin + 1]

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

ax.scatter(t_Par , bkgd_Par, color = "skyblue")
ax.set_title('1b) Points from {} to {} of random data'.format(t_Ini, t_Fin), fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background Distribution Value", fontsize = 20);
```

```python
Interval = 100
t_e100 = t[::Interval]
bkgd_e100 = bkgd[::Interval]

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

ax.scatter(t_e100, bkgd_e100, color = "pink")
ax.set_title('1c) Every {} points of random data'.format(Interval), fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background Distribution Value", fontsize = 20);
```

### Problem 2.

```python
bin_Size = 60

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

h = ax.hist2d(t, bkgd, bins = bin_Size, cmap = 'inferno')
ax.set_title('2a) Density plot of bin size {}'.format(bin_Size), fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Density", fontsize = 20);
fig.colorbar(h[3], ax=ax);
```

#### 2b)  
The plot is showing the density in the background distribution data.  
The brighter and redder the color is, the higher the density,  
while black represent a density of near 0.

### Problem 3.

```python
t_fold= t % (2* np.pi)
```

```python
plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

ax.scatter(t_fold, bkgd, color = "blue", s = 2)
ax.set_title('3a) Folded plot of random data', fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Background Distribution Value", fontsize = 20);
```

```python
bin_Size = 40

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')
fig, ax = plt.subplots(1, 1)

h = ax.hist2d(t_fold, bkgd, bins = bin_Size, cmap = 'viridis')
ax.set_title('3b) Density plot of bin size {} of folded data'.format(bin_Size), fontsize = 20)
ax.set_xlabel("Time", fontsize = 20)
ax.set_ylabel("Density", fontsize = 20);
fig.colorbar(h[3], ax=ax);
```
