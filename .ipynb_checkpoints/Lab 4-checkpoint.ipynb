#### 2)  
We know the background distribution is a Poisson distribution,  
which has a mean that varied due to the earth longitude.  
Next step is to find the variation.

```python
mean = np.zeros([2,36])
for i,val in enumerate(e_Long2):
    k = int(val/10)
    mean[0][k] +=1
    mean[1][k] += p_Count2[i]
for i in range(36):
    mean[1][i] /= mean[0][i]
```

```python
plt.rcParams["figure.figsize"] = (20,5)

fig, ax = plt.subplots(1,1)
bin_Size = 60
x_0 = np.linspace(5,355,36)
h0 = ax.hist2d(e_Long2, p_Count2,bins = bin_Size, cmap = 'GnBu')
ax.plot(x_0,mean[1], linewidth = 3, color = "Red")
ax.set_title('Earth Longitude vs Particle Count', fontsize = 20)
ax.set_xlabel("Earth Longitude (deg)", fontsize = 20)
ax.set_ylabel("Particle Count", fontsize = 20);
fig.colorbar(h0[3], ax=ax);
```

The variance is mark in red. It is a exponential decay function, in the form
$$
V = ab^x +c
$$
Time to do more plot to find the best fit exponential decay function:

```python
# reorganize the mean for better fitting:
y = np.zeros([36])
for i, val in enumerate(mean[1]):
    y[(i+5)%36] = val
```

```python
def exp_decay(x_0,a,b,c):
    return a*b**(x_0)+c

def fit_exp_nonlinear(x, y):
    opt_parms, parm_cov = scipy.optimize.curve_fit(exp_decay, x, y, maxfev=100000)
    a,b,c = opt_parms
    return a,b,c

a,b,c = fit_exp_nonlinear(x_0,y)
fit_y = exp_decay(x_0, a, b, c)
```

```python
plt.rcParams["figure.figsize"] = (20,5)
fig, ax = plt.subplots(1,1)
bin_Size = 60

ax.plot(x_0,y,'r', linewidth = 3, label = "data mean (shifted)")
ax.plot(x_0,fit_y, linewidth = 3, label = "fit")
ax.set_title('Background variance fit (Shifted)', fontsize = 20)
ax.set_xlabel("Earth Longitude (deg)", fontsize = 20)
ax.set_ylabel("Particle Count", fontsize = 20);
ax.set_xlim(0,360)
ax.legend();
```

```python
plt.rcParams["figure.figsize"] = (20,5)
fig, ax = plt.subplots(1,1)
bin_Size = 60

ax.plot(x_0,mean[1],'r', linewidth = 3, label = "data mean")
ax.plot(x_0,exp_decay((x_0+40)%360, a, b, c), linewidth = 3, label = "fit")
ax.set_title('Background variance fit (orginal))', fontsize = 20)
ax.set_xlabel("Earth Longitude (deg)", fontsize = 20)
ax.set_ylabel("Particle Count", fontsize = 20);
ax.set_xlim(0,360)
ax.legend();
```

```python
a,b,c
```

so the variance should be:
$$
V = 4.7495* 0.9924^ x +5.3316
$$
where x is given by $ (E_{Long} + 40)\% 360$

```python
def variance(x):
    x_0 = (x +40) % 360
    return  exp_decay(x_0, a, b, c)
```

#### 3)  
Then, as we expect the background, a distribution  
with discrete value, a Poisson distribution, the model should be Poisson with variance  
equals to our variance function.  
$$
\!f(k, V(x))= \frac{V(x)^k e^{-V(x)}}{k!}
$$

#### 4)  
at Earth Longitude of 30 degree, $5 \sigma$ =  

```python
five_Sig = stats.norm.cdf(5)
stats.poisson.ppf(five_Sig, variance(30))
```

at Earth Longitude of 300 degree, $5 \sigma$ =  

```python
stats.poisson.ppf(five_Sig, variance(300))
```

at Earth Longitude of 350 degree, $5 \sigma$ =  

```python
stats.poisson.ppf(five_Sig, variance(350))
```

```python
fsig_Val = np.zeros(36)
for i,val in enumerate(x_0):
    fsig_Val[i] = (stats.poisson.ppf(five_Sig, variance(val)))
```

```python
plt.rcParams["figure.figsize"] = (20,5)
fig, ax = plt.subplots(1,1)

ax.plot(x_0,fsig_Val, "ro")
ax.set_title(r'5 $\sigma$ threshold for a 100 millisecond GRB at different times', fontsize = 20)
ax.set_xlabel("Earth Longitude (deg)", fontsize = 20)
ax.set_ylabel(r"5 $\sigma$ value", fontsize = 20);
```
