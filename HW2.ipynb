### 1)

The probability of rolling two ones is $ \frac{1}{6} * \frac{1}{6} = \frac{1}{36}$,  
while rolling sevens have the following combinations:  
1. one + six
2. two + five
3. three + four  

With each having a probability of:  
 $ \frac{1}{6} ^2 *2 = \frac{1}{18}$   
The total probability is $\frac{1}{18} * 3 = \frac{1}{6}$  


Thus it is more likely to roll sevens, by a ratio of $\frac{\frac{1}{36}}{{\frac{1}{6}}} = \frac{1}{6}$

```python
fig, ax = plt.subplots(2, 2)
x = np.linspace(2.95,3.05,1000)


ax[0,0].plot(x,stats.norm.pdf(x,loc = 3., scale = 0.01))
ax[0,0].set_title("One Gaussian")

ax[0,1].plot(x,stats.norm.pdf(x,loc = 3., scale = 0.01)*2)
ax[0,1].set_title("Two Gaussian")

ax[1,0].plot(x,stats.norm.pdf(x,loc = 3., scale = 0.01)*3)
ax[1,0].set_title("Three Gaussian")

ax[1,1].plot(x,stats.norm.pdf(x,loc = 3., scale = 0.01)*4)
ax[1,1].set_title("Four Gaussian")

plt.show()
```

The standard deviation of the resulting sum or average stays the same.
